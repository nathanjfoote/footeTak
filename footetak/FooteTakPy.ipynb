```python
import gym
import foote_tak
import numpy as np
```

```python
def printaction(action):
    
    if action[0] == 2:   
        
        _, space, type_, _ = action
        
        if (type_ == 1) or (type_ == 2):
            type_ = "Flat Stone"
            
        else:
            type_ = "Standing Stone"
        
        print("Place a " + type_ + ' into: ' + str(space))
    
    else:
        _, move_from, move_to, trail, _ = action
        
        count = 0
        
        for i in trail:
            count += i
        
        piece = ' piece'
        if count > 1:
            piece = ' pieces'
        
        print("Move " + str(count) + str(piece) + \
              ' from: ' + str(move_from) +\
              ' to: ' + str(move_to))
```

```python
env_play = gym.make('Tak-v0')
env_render = gym.make('Tak-v0', render_mode="human")

env_play.reset()
env_render.reset()
terminated = False
turn = 0

env_play.reset()
env_render.reset()
terminated = False

while not terminated:

    if terminated:
        break

    turn += 1

    if env_play.state['Player: '] == 1:
        player = 'Player 1'
    else:
        player = 'Player 2'

    print(f"Turn {turn}: {player} to move\n")

    rand_action = np.random.randint(len(env_play.get_valid_actions())) 

    discrete_action = env_play.get_valid_actions()[rand_action][-1]

    action = env_play.map_to_action(discrete_action)
    printaction(action)

    print(str(action) + '\n')

    observation, reward_collected, terminated, info = env_play.step(discrete_action)
    env_render.step(discrete_action)

    print(env_play.state["Board: "])
    print('\nEnd of turn \n')
    print("GAME OVER\n")

    if env_play.state["Player: "] == 1:
        if reward_collected == 1:
            print("Player 1 Wins")
        elif reward_collected == -1:
            print("Player 2 Wins")
        else:
            print("Draw!")
    elif env_play.state["Player: "] == -1:
        if reward_collected == 1:
            print("Player 2 Wins")
        elif reward_collected == -1:
            print("Player 1 Wins")
        else:
            print("Draw!")
    
```

```python
time = env_play.return_time()
f_to_total_time = []
total_time = 0

for f, stats in time.items():
    f_time = stats['total_time'] #*stats['call_count']
    f_to_total_time.append((f, f_time))
    total_time += f_time

sortedtimes = sorted(f_to_total_time, key=lambda x: x[1])
print(f"Total time for N games: {total_time} seconds. Each function is below.")
for f_to_time in sortedtimes:
    print(f"{f_to_time[0]}: {f_to_time[1]} seconds\n")
```
